```python
nut_butter={"Almond butter":10
           ,"Peanut Butter":8,
           "cashew butter":12,
           "vagitables":20
           }
fig,ax=plt.subplots()
ax.bar(nut_butter.keys(),nut_butter.values())
ax.set(title="butter and their price",ylabel="price(R)")

```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter.keys()),list(nut_butter.values()))
ax.set(title="butter and their price",xlabel="price(R)")
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

```python
#two options subplot
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter.keys(),nut_butter.values())
ax4.hist(np.random.randn(100))
```

```python
#subploting 
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
ax[0,0].plot(np.random.randn(100))
ax[0,1].scatter(np.random.random(100),np.random.random(100))
ax[1,1].bar(nut_butter.keys(),nut_butter.values())
ax[1,0].hist(np.random.randn(100))
```
